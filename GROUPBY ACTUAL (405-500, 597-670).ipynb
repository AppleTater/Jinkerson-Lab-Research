## Results of GroupBy mean, median, and count. Then reorder by descending value of 'Ratio Median'

```python
data = pd.read_csv('TAG Data.csv')

data = data[['gene' , 'R4-S#08-HLG_405-500_597-670_normalized_reads' , 'R4-S#08-ALL_405-500_597-670_normalized_reads']]
data ['Ratios: HLG / ALL (405-500,597-670)'] = data ['R4-S#08-HLG_405-500_597-670_normalized_reads'] / data['R4-S#08-ALL_405-500_597-670_normalized_reads']

data = data.sort_values('gene')
data = data.replace([np.inf, -np.inf], np.nan)
data = data.dropna()

data = data.drop(['R4-S#08-HLG_405-500_597-670_normalized_reads' , 'R4-S#08-ALL_405-500_597-670_normalized_reads'], axis = 1)

data = pd.DataFrame(data)

mean = data.groupby(['gene']).mean()
mean.rename(columns = {'Ratios: HLG / ALL (405-500,597-670)': 'Ratio Mean'}, inplace = True)
# mean

median = data.groupby(['gene']).median()
median.rename(columns = {'Ratios: HLG / ALL (405-500,597-670)': 'Ratio Median'}, inplace = True)
# median

count = data.groupby(['gene']).count()
count.rename(columns = {'Ratios: HLG / ALL (405-500,597-670)': 'Count'}, inplace = True)
# count

result = data.reset_index(drop = True)
result = pd.concat([mean, median, count], axis = 1)

result = result.sort_values(by = ['Ratio Median'], ascending = False)

result
```

## As the 'Ratio Median' decreases, genes with 'Counts'   â‰¥ 3 are left

```python
data = pd.read_csv('TAG Data.csv')

data = data[['gene' , 'R4-S#08-HLG_405-500_597-670_normalized_reads' , 'R4-S#08-ALL_405-500_597-670_normalized_reads']]
data ['Ratios: HLG / ALL (405-500,597-670)'] = data ['R4-S#08-HLG_405-500_597-670_normalized_reads'] / data['R4-S#08-ALL_405-500_597-670_normalized_reads']

data = data.sort_values('gene')
data = data.replace([np.inf, -np.inf], np.nan)
data = data.dropna()

data = data.drop(['R4-S#08-HLG_405-500_597-670_normalized_reads' , 'R4-S#08-ALL_405-500_597-670_normalized_reads'], axis = 1)

data = pd.DataFrame(data)

mean = data.groupby(['gene']).mean()
mean.rename(columns = {'Ratios: HLG / ALL (405-500,597-670)': 'Ratio Mean'}, inplace = True)
# mean

median = data.groupby(['gene']).median()
median.rename(columns = {'Ratios: HLG / ALL (405-500,597-670)': 'Ratio Median'}, inplace = True)
# median

count = data.groupby(['gene']).count()
count.rename(columns = {'Ratios: HLG / ALL (405-500,597-670)': 'Count'}, inplace = True)
# count

result = data.reset_index(drop = True)
result = pd.concat([mean, median, count], axis = 1)

greater = result[result['Count'] < 3].index
result.drop(greater, inplace = True)

result = result.sort_values(by = ['Ratio Median'], ascending = False)

result
```
