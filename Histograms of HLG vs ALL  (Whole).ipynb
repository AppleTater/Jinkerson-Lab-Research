### Every row which has "NaN" means that the ratio of HLG / ALL is not between 0 and 1, for not enriched

```python
data = pd.read_csv('TAG Data.csv')

data = data[['IB' , 'gene' , 'R5-S#11-HLG_WHOLE_normalized_reads' , 'R5-S#11-ALL_WHOLE_normalized_reads']]
data ['Ratios: HLG / ALL (Whole)'] = data ['R5-S#11-HLG_WHOLE_normalized_reads'] / data['R5-S#11-ALL_WHOLE_normalized_reads']

# ~ 0 for not enriched
filter = data['Ratios: HLG / ALL (Whole)'] > 0
filter1 = data ['Ratios: HLG / ALL (Whole)'] < 1
data.where(filter & filter1, inplace = True)

data = data.dropna()

histo = data['Ratios: HLG / ALL (Whole)'].hist(bins = 100, color = 'plum', grid = True, xlabelsize = 12, ylabelsize = 12)
plt.xlabel("Ratio of HLG / ALL (Whole)", fontsize = 15)
plt.ylabel("Frequency", fontsize = 15)

print histo
```

### Every row which has "NaN" means that the ratio of HLG / ALL is not  > 1, for enriched

```python
data = pd.read_csv('TAG Data.csv')

data = data[['IB' , 'gene' , 'R5-S#11-HLG_WHOLE_normalized_reads' , 'R5-S#11-ALL_WHOLE_normalized_reads']]
data ['Ratios: HLG / ALL (Whole)'] = data ['R5-S#11-HLG_WHOLE_normalized_reads'] / data['R5-S#11-ALL_WHOLE_normalized_reads']

# > 1 for enriched
filter = data['Ratios: HLG / ALL (Whole)'] > 1
data.where(filter, inplace = True)

newdata = data.replace([np.inf, -np.inf], np.nan)
newdata = newdata.dropna()

histo = newdata['Ratios: HLG / ALL (Whole)'].hist(bins = 100, color = 'skyblue', grid = True, xlabelsize = 12, ylabelsize = 12)
plt.xlabel("Ratio of HLG / ALL (Whole)", fontsize = 15)
plt.ylabel("Frequency", fontsize = 15)

print histo
```
